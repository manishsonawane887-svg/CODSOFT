```python
print("🏁 Welcome to the F1 Ticket Chatbot!")
print("You can ask about tickets, prices, locations, or booking.")
print("Type 'bye' to exit.\n")

while True:
    user = input("You: ").lower()

    if "hello" in user or "hi" in user:
        print("Bot: Hello! Ready for the Formula 1 race experience? 🏎️")

    elif "ticket" in user and "price" in user:
        print("Bot: F1 ticket prices usually range from $100 to $2000 depending on seating and race location.")

    elif "book" in user or "buy ticket" in user:
        print("Bot: You can book F1 tickets from the official Formula 1 website or authorized ticket sellers.")

    elif "race" in user or "location" in user:
        print("Bot: Formula 1 races take place in different countries like Monaco, Singapore, Abu Dhabi, and more.")

    elif "schedule" in user or "next race" in user:
        print("Bot: The F1 season usually runs from March to December with races almost every two weeks.")

    elif "best seat" in user:
        print("Bot: Grandstand seats near the start/finish line offer the best race view!")

    elif "bye" in user or "exit" in user:
        print("Bot: Goodbye! Enjoy the Formula 1 season 🏁")
        break

    else:
        print("Bot: Sorry, I didn’t understand that. Try asking about F1 tickets, prices, or race locations.")
```
