```python
import pandas as pd

data = {
    "Movie": [
        "Inception", "Interstellar", "The Dark Knight",
        "Titanic", "The Notebook",
        "Avengers", "Iron Man", "Doctor Strange",
        "The Conjuring", "Insidious"
    ],
    "Genre": [
        "Sci-Fi", "Sci-Fi", "Action",
        "Romance", "Romance",
        "Action", "Action", "Fantasy",
        "Horror", "Horror"
    ]
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
genre_matrix = vectorizer.fit_transform(df["Genre"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(genre_matrix)
```

```python
def recommend(movie_name):
    movie_index = df[df["Movie"] == movie_name].index[0]

    scores = list(enumerate(similarity[movie_index]))

    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print("Recommended Movies:")

    for i in sorted_scores[1:4]:
        print(df.iloc[i[0]].Movie)
```

```python
recommend("Inception")
```
